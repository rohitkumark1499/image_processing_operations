# Sharping Image

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

original_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/colour_image.png',cv2.IMREAD_COLOR)


kernel = np.array([[0,-1,0],
                   [-1,5,-1],
                   [0,-1,0]])

sharpned_image = cv2.filter2D(original_image, -1, kernel)

print("original_image")
cv2_imshow(original_image)

print("sharpned_image")
cv2_imshow(sharpned_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
