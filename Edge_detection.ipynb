# **Edge_detection**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

original_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/colour_image.png')

#convert color to gray scale image
gray_scale_image = cv2.cvtColor(original_image,cv2.COLOR_BGR2GRAY)

#or we can convert colour image to gray scale image while reading it in imread()
# gray_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/colour_image.png',cv2.IMREAD_GRAYSCALE)

kernel = np.array([[0,1,0],
                   [1,-4,1],
                   [0,1,0]])


result_image = cv2.filter2D(gray_scale_image, -1, kernel)

#the above kernel is also known as laplacian kernal
result_image_without_kernal = cv2.Laplacian(gray_scale_image,-1)

print("original_image")
cv2_imshow(original_image)

print("gray_scale_image")
cv2_imshow(gray_scale_image)

# print("gray_image")
# cv2_imshow(gray_image)

print("result_image")
cv2_imshow(result_image)

print("result_image_without_kernal")
cv2_imshow(result_image_without_kernal)


cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Vertical Edge Detection**


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

original_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/building.jpg',cv2.IMREAD_GRAYSCALE)


kernel = np.array([[-1,0,1],
                   [-1,0,1],
                   [-1,0,1]])

result_image = cv2.filter2D(original_image, -1, kernel)

print("original_image")
cv2_imshow(original_image)

print("result_image")
cv2_imshow(result_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

# **Horizontal Edge Detection**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

original_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/building.jpg',cv2.IMREAD_GRAYSCALE)


kernel = np.array([[-1,-1,-1],
                   [0,0,0],
                   [1,1,1]])

result_image = cv2.filter2D(original_image, -1, kernel)

print("original_image")
cv2_imshow(original_image)

print("result_image")
cv2_imshow(result_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
