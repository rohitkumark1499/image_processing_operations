# **Blur Operation On Image**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

original_image = cv2.imread('/content/drive/MyDrive/Computer_vision/Images/colour_image.png',cv2.IMREAD_COLOR)

kernal = np.ones((5,5)) / 25
print(kernal)

blur_image = cv2.filter2D(original_image,-1,kernel=kernal)

print("original_image")
cv2_imshow(original_image)

print("blur_image")
cv2_imshow(blur_image)
```
